```python
docs_dataset = pd.DataFrame([UK,rishi_sunak, india])
docs_dataset.columns = ['Documents']
docs_dataset
```

```python
def punctuation_removal(txt):
  notations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
  for words in txt:
    if words in notations:
      txt = txt.replace(words, "")
  return txt.lower()
docs_dataset['punct_remove'] = docs_dataset['Documents'].apply(lambda y:punctuation_removal(y))

docs_dataset['punct_remove'] = docs_dataset['punct_remove'].apply(lambda x: w_tokenizer.tokenize(x))

def stopwords_removal(token_data):
  text =token_data
  output=[]
  for i in text:
    ft=remove_stopwords(i)
    if ft == '':
      continue 
    else :
      output.append(ft)
  return output

docs_dataset['final_doc']=docs_dataset['punct_remove'].apply(lambda x:stopwords_removal(x))
docs_dataset=docs_dataset.drop(columns=['punct_remove'],axis=1)
docs_dataset
```

```python
total_words=[]
for i in range(3):
  tokens=docs_dataset.iloc[i]["final_doc"]
  total_words.append(tokens)
print(total_words)
```

```python
#finding Unique words in all  the documents
def find_unique(tokens_list):
  n=len(tokens_list)
  s=1
  wordset=set(tokens_list[0])
  while s<n:
    uniq_wordset=wordset.union(set(tokens_list[s]))
    s=s+1
  return uniq_wordset


# set of unique words
unique_words=find_unique(total_words)
print(unique_words)
```

```python
def calculateBOW(wordset,doc):
  tf_diz = dict.fromkeys(wordset,0)
  for word in doc:
      tf_diz[word]=int(bool(doc.count(word)))

  return tf_diz
```

```python
Bag_of_words=[]
for doc in total_words:
    Bag_of_words.append(calculateBOW(unique_words,doc))
Bag_of_words
```

```python
df_bow_all = pd.DataFrame(Bag_of_words)
df_bow_all = df_bow_all.fillna(0)
df_bow_all = df_bow_all.T.astype(int)
df_bow_all
```

```python
df_bow_all['words'] = df_bow_all.index
word_tab=df_bow_all.reset_index()
word_tab=word_tab.drop(columns=['index'],axis=1)
word_tab
```
